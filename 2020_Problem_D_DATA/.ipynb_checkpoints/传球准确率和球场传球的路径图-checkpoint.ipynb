## 1.看看传球准确率

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
fullevents = pd.read_csv("fullevents.csv")
passingevents = pd.read_csv("passingevents.csv")
matches = pd.read_csv("matches.csv")
```

```python
fullevents[fullevents["EventType"]=="Pass"]["EventSubType"].unique()==passingevents["EventSubType"].unique()
```

```python
# 判断x,y所属的区域，给格子赋予0-100的编号
## 编码方式是从下到上1-10,11-20
def section(x,y):
    if(x<100 and y<100):
        return (np.floor(x/10.0)+1)+10*np.floor(y/10.0)
    elif(x==100 and y<100):
        return np.floor(x/10.0)+10*np.floor(y/10.0)
    elif(x<100 and y==100):
        return np.floor(x/10.0)+10*(np.floor(y/10.0)-1)
    elif(x==100 and y==100):
        return 100
```

```python
# 这个没用了..
#添加Section到新的表fullevents_Section中
fullevents_Section=fullevents.copy()

x=fullevents["EventOrigin_x"].values
y=fullevents["EventOrigin_y"].values
Section=[]
for i in range(0,len(x)):
    Section.append(section(x[i],y[i]))
fullevents_Section["OrgSection"]=Section

x=fullevents["EventDestination_x"].values
y=fullevents["EventDestination_y"].values
Section=[]
for i in range(0,len(x)):
    Section.append(section(x[i],y[i]))
fullevents_Section["DstSection"]=Section
```

```python
#查看一下数据
fullevents_Section.head()
```

```python
#选出我们要研究的数据，以"Pass"为例，存储到TargetData
TargetData=fullevents_Section[fullevents_Section["EventType"]=="Pass"].reset_index(drop=True)
TargetData.head()

# 可以研究一下Pass的子类别['Head pass', 'Simple pass', 'Launch', 'High pass', 'Hand pass','Smart pass', 'Cross']
# TargetData=fullevents_Section[fullevents_Section["EventSubType"]=="Simple pass"].reset_index(drop=True)
# TargetData.head()
```

```python
#这里用DestinationPlayerID是否为空添加一列判断传球是否成功,1表示失败！！
Fail=pd.isnull(TargetData["DestinationPlayerID"]).astype(int)
TargetData["Fail"]=Fail 
TargetData.head()
```

```python
#查询某场比赛的起始和结束的Index
#要求数据按照matchID升序排列,只能查询单场比赛数据
def getIndex(data,matchID):    
    i=0
    while data.loc[i,"MatchID"]!=matchID:
        i+=1
    start=i
    
    while i<data.shape[0] and data.loc[i,"MatchID"]==matchID:
        i+=1
    end=i
    return start,end

# 找到当前比赛的对手：
# 只能查询单场比赛数据
def getOpponent(data,matchID):
    start,end=getIndex(data,matchID)
    for i in range(start,end):
        if data.loc[i,"TeamID"]!='Huskies':
            return data.loc[i,"TeamID"]
```

```python
#查询某场比赛某队的失误率：
#判断依据是Data最后一列的Fail是否为1，1即为失误
#team=0本队，team=1 当前比赛的对手队
def FindError(data,matchId,team):
    start,end=getIndex(data,matchId)
    if(team==0):
        target="Huskies"
    elif(team==1):
        target=getOpponent(data,matchId)
    temp=data[start:end].reset_index(drop=True)
    return temp[temp["TeamID"]==target]["Fail"].mean()
```

```python
#统计Huskies和Opponent各场比赛的传球成功率：
#一个偷懒的写法..时间复杂度比较高..
Huskies=[]
Opponent=[]
for i in range(1,39):
    Huskies.append(1-FindError(TargetData,i,0))
    Opponent.append(1-FindError(TargetData,i,1))
```

```python
# 处理一下比赛数据
# score = OwnScore-OpponentScore,-2表示Huskies输了两球
score = matches["OwnScore"].values-matches["OpponentScore"].values
matches_score=matches.copy()
matches_score["Score"]=score
# matches_score.head()
Huskies_index=[] #Huskies赢的场数
Huskies_size=[] #Huskies赢的比分,1->size,100
Opponent_index=[]
Opponent_size=[]
Tie_index=[] #平局默认size=50
win=[] #统计Huskies是否赢了比赛,1表示赢了，0表示平了，-1表示输了

for i in range(0,38):
    if(matches_score["Score"][i]>0):
        Huskies_index.append(i)
        Huskies_size.append(100*matches_score["Score"][i])
        win.append(1)
        
    elif(matches_score["Score"][i]<0):
        Opponent_index.append(i)
        Opponent_size.append(abs(100*matches_score["Score"][i]))
        win.append(0)
        
    else:
        Tie_index.append(i)
        win.append(-1)
```

```python
fig = plt.figure(figsize=(10,6))
plt.plot(range(1,39),Opponent,'o-',label="Opponent")
plt.plot(range(1,39),Huskies,'x-',label="Huskies")
# plt.scatter([1,2],Huskies[0:2],marker='o',s=[100,50],c="yellow",label="Huskise win")
plt.scatter([i+1 for i in Huskies_index],[Huskies[i]for i in Huskies_index],marker='o',s=Huskies_size,c="red",label="win")
plt.scatter([i+1 for i in Opponent_index],[Opponent[i]for i in Opponent_index],marker='o',s=Opponent_size,c="red")
plt.scatter([i+1 for i in Tie_index],[Opponent[i]for i in Tie_index],marker='o',s=50,c="green",label="Tie") #平局的情况
plt.scatter([i+1 for i in Tie_index],[Huskies[i]for i in Tie_index],marker='o',s=50,c="green")
plt.xlabel("MatchID")
plt.ylabel("Passing Accuracy")
plt.legend()
plt.show()
```

```python
threshold=0.05
compare=[] #用于比较根据accuracy差值小于threshold为平，其余哪个大就是哪队赢的预测和真实比赛胜负情况
for i in range(0,38):
    err=(Huskies[i]-Opponent[i])
    if win[i]==0:
        if np.abs(err)<=threshold:
            compare.append(1)
        else:
            compare.append(0)
            print("第%d场比赛 真实是Tie但预测错了" %(i+1))
    elif win[i]==1:
        if err>0 and np.abs(err)>threshold:
            compare.append(1)
        else:
            compare.append(0)
            print("第%d场比赛 真实是Win预测错了" %(i+1))
    elif win[i]==-1:
        if err<0 and np.abs(err)>threshold:
            compare.append(1)
        else:
            compare.append(0)
            print("第%d场比赛 真实是Loss预测错了" %(i+1))
print("Accuracy=",np.mean(compare))
```

##  2.统计球场传球的路径图

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
fullevents = pd.read_csv("fullevents.csv")
passingevents = pd.read_csv("passingevents.csv")
matches = pd.read_csv("matches.csv")
```

```python
# 判断x,y所属的区域，给格子赋予0-100的编号
## 编码方式是从下到上1-10,11-20
def section(x,y):
    if(x<100 and y<100):
        return (np.floor(x/10.0))+10*np.floor(y/10.0)
    elif(x==100 and y<100):
        return np.floor(x/10.0)+10*np.floor(y/10.0)-1
    elif(x<100 and y==100):
        return np.floor(x/10.0)+10*(np.floor(y/10.0)-1)
    elif(x==100 and y==100):
        return 99
```

```python
# 这个没用了..
#添加Section到新的表fullevents_Section中
fullevents_Section=fullevents.copy()

x=fullevents["EventOrigin_x"].values
y=fullevents["EventOrigin_y"].values
Section=[]
for i in range(0,len(x)):
    Section.append(section(x[i],y[i]))
fullevents_Section["OrgSection"]=Section
fullevents_Section["OrgSection"]=fullevents_Section["OrgSection"].values.astype(int)

x=fullevents["EventDestination_x"].values
y=fullevents["EventDestination_y"].values
Section=[]
for i in range(0,len(x)):
    Section.append(section(x[i],y[i]))
fullevents_Section["DstSection"]=Section
fullevents_Section["DstSection"]=fullevents_Section["DstSection"].values.astype(int)
fullevents_Section.head()
```

```python
# 我发现还是用(x,y)定义格子方便一点...重新来过，左闭右开
def FindAbscisse(x):
    if x==100:
        return 10
    else:
        return x//10+1
fullevents_Section["OrgX"]=fullevents["EventOrigin_x"].apply(FindAbscisse).values.astype(int)
fullevents_Section["OrgY"]=fullevents["EventOrigin_y"].apply(FindAbscisse).values.astype(int)
fullevents_Section["DstX"]=fullevents["EventDestination_x"].apply(FindAbscisse).values.astype(int)
fullevents_Section["DstY"]=fullevents["EventDestination_y"].apply(FindAbscisse).values.astype(int)
fullevents_Section.head()
```

```python
TargetData=fullevents_Section[fullevents_Section["EventType"]=="Pass"].reset_index(drop=True)
#这里用DestinationPlayerID是否为空添加一列判断传球是否成功,1表示失败！！
Fail=pd.isnull(TargetData["DestinationPlayerID"]).astype(int)
TargetData["Fail"]=Fail 
TargetData.head()
```

```python
# 构建有向的传球网络
# index是自己选择的指标可以是一场比赛或者是上半场下面场等等
index=(TargetData["MatchID"]==1).values*(TargetData["TeamID"]=="Huskies").values
# index=(TargetData["MatchID"]==1).values*(TargetData["OriginPlayerID"]=="Opponent14_M4").values

def ConstructPath(index,data):
    matrix=np.zeros([100,100])
    temp=data[index].reset_index(drop=True)
    for i in range(temp.shape[0]):
        matrix[temp["OrgSection"][i],temp["DstSection"][i]] +=1
    return matrix
matrix=ConstructPath(index,TargetData)
```
